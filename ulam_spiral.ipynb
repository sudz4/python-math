* The Ulam spiral, also known as the Ulam spiral of prime numbers, is a visual representation of prime numbers, discovered by the mathematician Stanislaw Ulam in 1963. The spiral arranges integers on a grid in a spiral pattern, starting with the number 1 at the center, and then moving outward in a counter-clockwise direction. The unique aspect of the Ulam spiral is that when the prime numbers on the grid are highlighted, they appear to form diagonal lines or other distinct patterns.

```python
import numpy as np
import matplotlib.pyplot as plt

def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def generate_ulam_spiral(size):
    x, y = 0, 0
    dx, dy = 0, -1
    count = 1
    coords = []

    for _ in range(size**2):
        if is_prime(count):
            coords.append((x, y))
        if x == y or (x < 0 and x == -y) or (x > 0 and x == 1 - y):
            dx, dy = -dy, dx
        x, y = x + dx, y + dy
        count += 1

    return coords

size = 201  # Choose an odd number for the size of the grid
coords = generate_ulam_spiral(size)
x, y = zip(*coords)

plt.figure(figsize=(10, 10))
plt.scatter(x, y, s=10, marker='s', edgecolors='none')
plt.title("Ulam Spiral")
plt.axis('equal')
plt.xticks(np.arange(-size // 2, size // 2, step=10))
plt.yticks(np.arange(-size // 2, size // 2, step=10))
plt.grid(True)
plt.show()

```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def generate_ulam_spiral(size):
    spiral = np.zeros((size, size), dtype=int)
    x, y = size // 2, size // 2
    dx, dy = 0, -1
    count = 1

    for _ in range(size**2):
        if (-size // 2 < x <= size // 2) and (-size // 2 < y <= size // 2):
            spiral[size // 2 - y][x + size // 2] = count if is_prime(count) else 0
        if x == y or (x < 0 and x == -y) or (x > 0 and x == 1 - y):
            dx, dy = -dy, dx
        x, y = x + dx, y + dy
        count += 1

    return spiral

size = 201  # Choose an odd number for the size of the grid
spiral = generate_ulam_spiral(size)

plt.figure(figsize=(10, 10))
plt.imshow(spiral, cmap="coolwarm", extent=(-size // 2, size // 2, -size // 2, size // 2))
plt.title("Ulam Spiral")
plt.show()
```
