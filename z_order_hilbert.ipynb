| Hilbert Curve | Also know as the Hilbert space-filling curve |
| ----------- | ----------- |

 Constructed by recursively subdividing a square or a cube into smaller squares or cubes and connecting their centers in a particular pattern

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def hilbert_curve(x,y,size,order):
    if size ==1:
        # return coordinates of single point in the curve
        return np.array([[x,y]])

        # recursively compute coordiantes of the four quadrants of the curve

    q1 = hilbert_curve(x, y, size//2, order-1) # // is the floor operator. keeps this organized as an itneger
    q2 = hilbert_curve(x + size//2, y, size//2, order-1)
    q3 = hilbert_curve(x + size//2, y + size//2, size//2, order-1)
    q4 = hilbert_curve(x, y + size//2, size//2, order-1)

    # Interleave the coordinates of the four quadrants to form complete curve
    curve = np.vstack((q1, q2, q3, q4))
    if order % 2 == 0:
        curve = curve[:, ::-1]

    return curve
```

 | "Interleaving" | Z-order indexing concept |
 |----------------|--------------------------|
 
 Interleaving coordinates refers to the process of combining the coordinates of multiple points in a multi-dimensional space in such a way that the resulting sequence preserves the spatial relationships of the points.

```python
# compute coordinates of the hilbert curve
size = 8
order = 3

curve = hilbert_curve(0, 0, size, order)
```

```python
# Plot the curve
plt.plot(curve[:, 0], curve[:, 1])
plt.show()
```
